```python
import pandas as pd 
import numpy as np
```

```python
books = pd.read_csv('Data/Books.csv')
ratings = pd.read_csv('Data/Ratings.csv')
```

```python
books.head()
```

```python
books.columns
```

```python
books = books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher',
       'Image-URL-L']]
```

```python
books.head()
```

```python
books.rename(columns={
    "Book-Title": "Title",
    "Book-Author": "Author",
    "Year-Of-Publication": "Year",
    "Image-URL-L": "URL"
}, inplace=True)
```

```python
books.head(3)
```

```python
ratings.head()
```

```python
ratings['User-ID'].value_counts()
```

```python
# filter the users who have rated less than 200 books 
X = ratings['User-ID'].value_counts() > 200
X
```

```python
y = X[X].index
y
```

```python
y.shape
```

```python
ratings = ratings[ratings['User-ID'].isin(y)]
ratings['User-ID'].value_counts()
```

```python
# merge the ratings with the books 
books_with_ratings = books.merge(ratings, on='ISBN')
books_with_ratings.head()
```

```python
# group the books with the ratings and find their counts 
num_ratings = books_with_ratings.groupby('Title')['Book-Rating'].count().reset_index()
```

```python
num_ratings['Book-Rating'].max()
```

```python
num_ratings.head()
```

```python
num_ratings.shape
```

```python
num_ratings.rename(columns={
    "Book-Rating": "Total-Rating"
}, inplace=True)
```

```python
num_ratings.sample()
```

```python
final_rating = books_with_ratings.merge(num_ratings, on='Title')
final_rating.head()
```

```python
final_rating = final_rating[final_rating['Total-Rating'] >= 50]
final_rating
```

```python
# create a pivot table 
pivot_table = final_rating.pivot_table(values='Book-Rating', columns='User-ID', index='Title')
```

```python
pivot_table
```

```python
pivot_table.fillna(0, inplace=True)
```

```python
pivot_table
```

```python
# create a sparse matrix
from scipy.sparse import csr_matrix

book_sparse = csr_matrix(pivot_table)
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestions = model.kneighbors(pivot_table.iloc[234,:].values.reshape(1,-1), n_neighbors=5)
```

```python
distance
```

```python
book_names = pivot_table.index
```

```python
# save the models 
import pickle
pickle.dump(model, open('models/model.pkl', 'wb'))
pickle.dump(book_names, open('models/book_names.pkl', 'wb'))
pickle.dump(pivot_table, open('models/book_pivot.pkl', 'wb'))
pickle.dump(final_rating, open('models/final_ratings.pkl', 'wb'))
```

```python
def recommmend_book(book_name):
    book_id = np.where(pivot_table.index == book_name)[0][0]
    distance, suggestions = model.kneighbors(pivot_table.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6)

    for i in range(len(suggestions)):
        books = pivot_table.index[suggestions[i]]
        for j in books:
            print(j)
```

```python
recommmend_book("Year of Wonders")
```

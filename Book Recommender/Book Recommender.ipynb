### Book Recommender System using Collaborative based

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
books = pd.read_csv('Data/Books.csv')
books.head()
```

```python
# columns 
books.columns
```

```python
# shape
books.shape
```

```python
# description
books.describe()
```

```python
books = books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Image-URL-L']]
books.head()
```

```python
# rename the columns 
books.rename(columns={
    "Book-Title": "Title",
    "Book-Author" : "Author",
    "Year-Of-Publication": "Year",
    "Image-URL-L": "URL"
}, inplace=True)
```

```python
books.head()
```

```python
# check for null values 
books.isnull().sum()
```

```python
# drop the null values 
books = books.dropna()
```

```python
books.isnull().sum()
```

```python
# load users 
users = pd.read_csv('Data/Users.csv')
users.head()
```

```python
users.shape
```

```python
# import the ratings
ratings = pd.read_csv('Data/Ratings.csv')
ratings.head()
```

```python
books.shape, users.shape, ratings.shape
```

```python
# renaming the rating columns 
ratings.rename(columns={
    "User-ID": "user_id",
    "Book-Rating": "rating"
}, inplace=True)
```

```python
ratings.head()
```

```python
ratings.user_id.value_counts()
```

```python
ratings['user_id'].unique().shape
```

```python
# remove the user who have not rated more than 200  books 
X = ratings['user_id'].value_counts() > 200 
```

```python
X[X].shape
```

```python
y = X[X].index
y.shape
```

```python
ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
books.head(2)
```

```python
ratings_with_books = ratings.merge(books, on="ISBN")
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
num_rating = ratings_with_books.groupby('Title')['rating'].count().reset_index()
```

```python
num_rating.head()
```

```python
num_rating.rename(columns={"rating": "num_of_ratings"}, inplace=True)
```

```python
num_rating.head()
```

```python
final_rating = ratings_with_books.merge(num_rating, on="Title")
```

```python
final_rating.sample()
```

```python
final_rating.shape
```

```python
# filtering 
final_rating = final_rating[final_rating['num_of_ratings']>=50]
```

```python
final_rating.sample(20)
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id', 'Title'], inplace=True)
```
